```python
# Importing pandas library for data manipulation
import pandas as pd

# Reading the medical insurance dataset from a CSV file
medical_dataset = pd.read_csv("medical_insurance.csv")
```

```python
# Importing necessary libraries for machine learning
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
```

```python
# Encoding categorical variables using LabelEncoder
SeX = LabelEncoder()

medical_dataset['sex'] = SeX.fit_transform(medical_dataset['sex'])

Smoker = LabelEncoder()

medical_dataset['smoker'] = Smoker.fit_transform(medical_dataset['smoker'])

Region = LabelEncoder()

medical_dataset['region'] = Region.fit_transform(medical_dataset['region'])

# Displaying the first few rows of the modified dataset
medical_dataset.head()
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(medical_dataset[['age','sex','bmi','children','smoker','region']].values,medical_dataset['charges'],test_size=0.2)
```

```python
# Instantiating the Linear Regression model
lr = LinearRegression()
# Fitting the model on the training data
lr.fit(X_train,y_train)
```

```python
# Obtaining input data from the user
Age = int(input("Enter your age: "))

Sex = int(input("sex: male=1,female=0: " ))

Bmi = float(input("Enter your BMI: "))

Children = int(input("how many children do you have: "))

Smoker = int(input('smoker: yes=1,no=0: ' ))

Region = int(input("region: southeast=2,southwest=3,northeast=0,northwest=1: "))

# Predicting medical charges based on user input
print("Your Medical Charges would be : $",lr.predict([[Age,Sex,Bmi,Children,Smoker,Region]]))
```
